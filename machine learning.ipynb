```python
import numpy as np
from scipy.optimize import linprog

c = np.array([5, 10, 8]) # coefficients for the objective function
A = np.array([[1, 1, 1], [3, 5, 2]]) # coefficients for the constraints
b = np.array([72, 60]) # right-hand side of the constraints
x0_bounds = (0, None) # lower and upper bounds for x0 (admin staff)
x1_bounds = (0, None) # lower and upper bounds for x1 (astronauts)
x2_bounds = (0, None) # lower and upper bounds for x2 (engineers)

# Use linprog from the scipy library to solve the optimization problem
res = linprog(c, A_ub=A, b_ub=b, bounds=[x0_bounds, x1_bounds, x2_bounds], method='simplex')

# Resulting solution
admin_staff = int(res.x[0])
astronauts = int(res.x[1])
engineers = int(res.x[2])

# The maximum funding
funding = 5 * admin_staff + 10 * astronauts + 8 * engineers

print("Admin Staff:", admin_staff)
print("Astronauts:", astronauts)
print("Engineers:", engineers)
print("Max Funding:", funding)
```
