```python
import requests

# API endpoint for Yelp Fusion API search
search_endpoint = "https://api.yelp.com/v3/businesses/search"

# API endpoint for Yelp Fusion API reviews
reviews_endpoint = "https://api.yelp.com/v3/businesses/{business_id}/reviews"

# API Key
api_key = "jqRzrGW_W55mstAZYzt7p8FRkfb3CkNpzhvvfh6aysjOI-jBPAFsPUKYOWEPMJ49mdkAKU-JTnM7ow7IVUhhg0T12o5aOvTl9tem3r3-mcggzXzms4sPBhFCWwjXY3Yx"

# Search parameters
params = {
    "term": "Hilton",
    "location": "Toronto",
    "limit": 50
}

# Set API headers
headers = {
    "Authorization": f"Bearer {api_key}"
}

# Make API request to search for hotels
response = requests.get(search_endpoint, headers=headers, params=params)

# Check if the API request was successful
if response.status_code == 200:
    # Get the business IDs from the API response
    businesses = response.json()["businesses"]
    business_ids = [business["id"] for business in businesses]
    
    # Keep track of the review texts and ratings
    review_texts = []
    ratings = []
    
    # Loop through each business ID
    for business_id in business_ids:
        # Make API request to get the reviews for a business
        response = requests.get(reviews_endpoint.format(business_id=business_id), headers=headers)
        
        # Check if the API request was successful
        if response.status_code == 200:
            # Get the review texts and ratings from the API response
            reviews = response.json()["reviews"]
            print(reviews)
            review_texts.extend([review["text"] for review in reviews])
            ratings.extend([review["rating"] for review in reviews])
            
            # Check if we have at least 1000 review texts and ratings
            if len(review_texts) >= 1000:
                break
        else:
            # Handle errors from the API request
            print(f"Reviews API request failed for business ID {business_id} with status code: {response.status_code}")
    
    print(f"Review Texts: {review_texts[:1000]}")
    print(f"Ratings: {ratings[:1000]}")
else:
    # Handle errors from the API request
    print(f"Search API request failed with status code: {response.status_code}")
```

```python
import pandas as pd
df = pd.DataFrame(review_texts,ratings)
```

```python
df.reset_index(inplace = True)
```
